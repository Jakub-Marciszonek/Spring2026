## Regression (Neural Network predicting a number)

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Synthetic linear data
X = np.arange(-20, 21, 2, dtype=np.float32) # [-20, -18, ..., 20]
y = X + 10
```

```python
# Train/test split (simple)
X_train, y_train = X[:20], y[:20]
X_test, y_test = X[20:], y[20:]
```

```python
# 1) CREATE
model = tf.keras.Sequential([
 tf.keras.layers.Input(shape=(1,)),
 tf.keras.layers.Dense(1) # linear output for regression
])
```

```python
# 2) COMPILE
model.compile(
 loss="mae",
 optimizer=tf.keras.optimizers.SGD(learning_rate=0.01), 
 # Changed learning rate from 0.01 --> 0.05
 metrics=["mae"]
)
```

```python
# 3) FIT
history = model.fit(X_train, y_train, epochs=200, verbose=1)
```

```python
# Evaluate
loss, mae = model.evaluate(X_test, y_test, verbose=1)
print("Test MAE:", mae)
```

```python
# Predict and visualize
y_pred = model.predict(X_test).squeeze()
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, c='b', label='Train')
plt.scatter(X_test, y_test, c='g', label='Test')
plt.scatter(X_test, y_pred, c='r', label='Pred')
plt.legend(); plt.title("Regression: y = x + 10 learned")
plt.show()
```

Current version is not doing very well most of the time MAE is around 10+ which is not excelent

```python
# Synthetic linear data
X = np.arange(-20, 21, 2, dtype=np.float32) # increased number of data
y = X + 10
# Train/test split (simple)
X_train, y_train = X[:20], y[:20]
X_test, y_test = X[20:], y[20:]
# 1) CREATE
model = tf.keras.Sequential([
 tf.keras.layers.Input(shape=(1,)),
 tf.keras.layers.Dense(1) # linear output for regression
])
# 2) COMPILE
model.compile(
 loss="mae",
 optimizer=tf.keras.optimizers.SGD(learning_rate=0.05), 
 # changed learning rate from 0.01 to 0.05
 metrics=["mae"]
)
# 3) FIT
history = model.fit(X_train, y_train, epochs=300, verbose=1) # increased number of epochs from 200 --> 300
# Evaluate
loss, mae = model.evaluate(X_test, y_test, verbose=1)
print("Test MAE:", mae)
# Predict and visualize
y_pred = model.predict(X_test).squeeze()
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, c='b', label='Train')
plt.scatter(X_test, y_test, c='g', label='Test')
plt.scatter(X_test, y_pred, c='r', label='Pred')
plt.legend(); plt.title("Regression: y = x + 10 learned")
plt.show()
```

After changing the learning rate from 0.01 to 0.05 MAE reduced to ~5 which is excelent considering that algorythm is not using more recources
